```python
from sklearn.linear_model import LinearRegression
import numpy as np
import matplotlib.pyplot as plt
```

```python
np.random.seed(42)
x = 10 + np.random.rand(50, 1)
y = 3*x + 10 + np.random.rand(50, 1)
```

```python
plt.scatter(x, y)
plt.xlabel("X")
plt.ylabel("Y")
plt.title("Scatter plot of X vs Y")
```

```python
model = LinearRegression()
history = model.fit(x, y)
```

```python
y_pred = model.predict(x)
```

```python
plt.scatter(x, y)
plt.plot(x, y_pred, color='red')
plt.xlabel("X")
plt.ylabel("Y")
plt.title("Data Points and Regression Line")
```

```python
m, c = history.intercept_, history.coef_
print(f"Model parameters: Intercept = {m}, Coefficient = {c}")
```
