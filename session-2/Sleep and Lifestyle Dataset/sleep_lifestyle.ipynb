# Initialize

```python
import pandas as pd
```

```python
df = pd.read_csv('sleep_lifestyle.csv')
```

```python
num_cols = [col for col in df.columns if df[col].dtype != 'object']
cat_cols = [col for col in df.columns if df[col].dtype == 'object']

print("Numerical Columns:", num_cols)
print("Categorical Columns:", cat_cols)
```

# Cleaning Dataframe

```python
df[cat_cols].nunique()
```

```python
df.isnull().sum()
```

```python
df_processed = df.drop(columns=['Person ID'])
df_processed.head()
```

```python
df_processed['Sleep Disorder'].fillna('No', inplace=True)
df_processed.info()
```

# Identifying Outliers

```python
import matplotlib.pyplot as plt

plt.subplot(2,1,1)
plt.boxplot(df_processed['Age'], vert=False)
plt.title('Age')
plt.grid(True)
plt.subplot(2,1,2)
plt.boxplot(df_processed['Sleep Duration'], vert=False)
plt.title('Sleep Duration')
plt.grid(True)
plt.show()
```

```python
plt.subplot(2,1,1)
plt.boxplot(df_processed['Physical Activity Level'], vert=False)
plt.title('Physical Activity Level')
plt.grid(True)
plt.subplot(2,1,2)
plt.boxplot(df_processed['Stress Level'], vert=False)
plt.title('Stress Level')
plt.grid(True)
plt.show()
```

```python
plt.subplot(2,1,1)
plt.boxplot(df_processed['Heart Rate'], vert=False)
plt.title('Heart Rate')
plt.grid(True)
plt.subplot(2,1,2)
plt.boxplot(df_processed['Daily Steps'], vert=False)
plt.title('Daily Steps')
plt.grid(True)
plt.show()
```

```python
outliers = df_processed[df_processed['Heart Rate'] > 78]
df_processed = df_processed[df_processed['Heart Rate'] <= 78]
```

# Processing to Export

```python
x = df_processed[['Gender', 'Age', 'Occupation', 'Sleep Duration', 'Quality of Sleep', 'Physical Activity Level', 'Stress Level',
                  'BMI Category','Blood Pressure', 'Heart Rate','Daily Steps']]
y = df_processed['Sleep Disorder']
```

```python
x.head()
```

```python
y.head()
```

```python
num_cols = [col for col in x.columns if x[col].dtype != 'object']
cat_cols = [col for col in x.columns if x[col].dtype == 'object']
```

# Min Max Scaling

```python
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()
x[num_cols] = scaler.fit_transform(x[num_cols])
x.head()
```

# Standardization

```python
from sklearn.preprocessing import StandardScaler

sscaler = StandardScaler()
standardized_data = sscaler.fit_transform(x[num_cols])

standardized_df = pd.DataFrame(standardized_data, columns=num_cols)

print("Standardized Data (mean=0, std=1):")
print(standardized_df)
```

# One Hot Encoding

```python
from sklearn.preprocessing import OneHotEncoder

encoder = OneHotEncoder(sparse_output=False, handle_unknown='ignore')

encoded_data = encoder.fit_transform(x[cat_cols])

encoded_columns = encoder.get_feature_names_out(cat_cols)

x_encoded = pd.DataFrame(encoded_data, columns=encoded_columns)
```

# Export

```python
x.head()
```

```python
x.to_csv('x.csv', index=False)
y.to_csv('y.csv', index=False)

import joblib
joblib.dump(scaler, 'scaler.save')
joblib.dump(encoder, 'encoder.pkl')
joblib.dump(sscaler, 'std_scaler.save')
```
